* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_rows', 100)
```

### ① 取出今天所有的發文

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text)
soup

for i in soup.find(class_='part_list_2').find_all('h3'):
    print(i.find(class_='date').text, i.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
time = []
category = []
title = []
for i in soup.find(class_='part_list_2').find_all('h3'):
    time.append(i.find(class_='date').text)
    category.append(i.find('em').text)
    title.append(i.find('a').text)
df = pd.DataFrame(list(zip(time, category, title)), columns=['time', 'category', 'title'])
df.sort_values(by=['category'], inplace=True)
df
```

### ③ 哪一個類別的文章最多

```python
df['category'].value_counts().index[0]
```
