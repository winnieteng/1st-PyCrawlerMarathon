```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import re
category = table.find_all('td', attrs={'style':'white-space:nowrap;'})
date = table.find_all('td', text=re.compile('2019/\d\d'))
value = table.find_all('td', text=re.compile('^\d{1,2}$|^\d{1,2}\.*\d{1,2}$'))
month = int(len(date)/len(category))
data = dict()
for i in range(len(category)):
    subdata = dict()
    for j in range(month):
        subdata[date[j+i*month].text] = value[j+i*month].text
    data[category[i].text] = subdata
print(data)
```
