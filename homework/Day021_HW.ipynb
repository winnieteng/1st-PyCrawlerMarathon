## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import re
from datetime import datetime, timedelta
import time
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')  
url = 'https://www.ettoday.net/news/news-list.htm'
browser.get(url)
```

```python
today = datetime.now()
year = datetime.strftime(today, '%Y')
month = datetime.strftime(today, '%#m')
day = datetime.strftime(today, '%d')
```

```python
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(year)
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value(month)
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(day)

browser.find_element_by_id('button').click()
```

```python
js="var q=document.documentElement.scrollTop=100000"  
for i in range(50):
    time.sleep(2)  
    browser.execute_script(js)  
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
todaystr = datetime.strftime(today, '%Y/%m/%d')
article = soup.find_all('div', class_='part_list_2')[0].find_all('h3')
for i in range(len(article)):
    if article[i].find('span').text[:10] == todaystr:
        print('{} {}'.format(article[i].find('span').text, article[i].find('a').text))
    else:
        break
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta

now = datetime.now()
before = datetime.now()- timedelta(hours=2)
for i in range(len(article)):
    time = article[i].find('span').text
    timedelta = datetime.strptime(time, '%Y/%m/%d %H:%M')
    if before < timedelta and timedelta < now:
        print('{} {}'.format(article[i].find('span').text, article[i].find('a').text))
    else:
        break
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime, timedelta
threed = (today - timedelta(days=3))
year = datetime.strftime(threed, '%Y')
month = datetime.strftime(threed, '%#m')
day = datetime.strftime(threed, '%d')
```

```python
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(year)
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value(month)
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(day)

browser.find_element_by_id('button').click()
```

```python
import time
js="var q=document.documentElement.scrollTop=100000"  
for i in range(100):
    time.sleep(2)  
    browser.execute_script(js)  
```

```python
article = soup.find_all('div', class_='part_list_2')[0].find_all('h3')
# article[0].find('span').text[:10] == '2020/03/23'
for i in range(len(article)):
    if (article[i].find('span').text[11:13] == '15') or (article[i].find('span').text[11:13] == '16') or (article[i].find('span').text[11:] == '17:00'):
        print('{} {}'.format(article[i].find('span').text, article[i].find('a').text))
    elif article[i].find('span').text[11:13] == '14':
        break
    else:
        continue
```
